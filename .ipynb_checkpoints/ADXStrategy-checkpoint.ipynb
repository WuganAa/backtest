```python
import datetime
import numpy as np
import pandas as pd
import queue

from Backtest.backtest import Backtest
from Backtest.data import JSONDataHandler
from ADXStrategy import ADXStrategy
```

```python
def run(config, window):
    events_queue = queue.Queue()
    data_handler = JSONDataHandler(
        config['csv_dir'], config['freq'], events_queue, config['tickers'],
        start_date=config['start_date'], end_date=config['end_date']
    )
    strategy = ADXStrategy(data_handler, events_queue, suggested_quantity = 1,
                           window = window)

    backtest = Backtest(config, events_queue, strategy,
                        data_handler= data_handler)

    results = backtest.start_trading(config)
    return backtest, results
```

### 样本内

```python
config = {
    "csv_dir": "F:/Python/backtest/ethusdt-trade.csv.2018-07-25.formatted",
    "out_dir": "F:/Python/backtest/backtest/results/ADXStrategy/in_sample",
    "title": "ADXStrategy",
    "save_plot": True,
    "save_tradelog": True,
    "start_date": pd.Timestamp("2018-07-25T00:00:00", tz = "UTC"),
    "end_date": pd.Timestamp("2018-07-25T04:20:00", tz = "UTC"),
    "equity": 500.0,
    "freq": 1,      # min
    "tickers": ['ETHUSDT']
}
```

```python
interval = np.array([5, 10, 12, 26, 30, 35, 45, 60, 72, 84, 96, 120, 252])
interval
```

```python
ans = []
```

```python
for i in range(len(interval)):
    window = interval[i]
    config['title'] = "MACDStrategy" + "_" + str(window)
    backtest, results = run(config, window = window)
    dict_ans = {
        "window": window, 
        "Sharpe Ratio": results['sharpe'], 
        "Total Returns": (results['cum_returns'][-1] - 1), 
        "Max Drawdown": (results["max_drawdown"] * 100.0), 
        "Max Drawdown Duration": (results['max_drawdown_duration']),
        "Trades": results['trade_info']['trading_num'],
        "Trade Winning": results['trade_info']['win_pct'],
        "Average Trade": results['trade_info']['avg_trd_pct'],
        "Average Win": results['trade_info']['avg_win_pct'],
        "Average Loss": results['trade_info']['avg_loss_pct'],
        "Best Trade": results['trade_info']['max_win_pct'],
        "Worst Trade": results['trade_info']['max_loss_pct'],
        "Worst Trade Date": results['trade_info']['max_loss_dt'],
        "Avg Days in Trade": results['trade_info']['avg_dit']

    }
    ans.append(dict_ans)
ans = pd.DataFrame(ans)
```

```python
ans.to_csv("F:/Python/backtest/backtest/results/ADXStrategy"  + "/result_ADXStrategy_in_sample.csv.csv")
```
